```python
def fill_cols(d, tempd):
    for keys in d:
        try:
            if(tempd[keys]):
                d[keys] = d[keys].append({keys: tempd[keys]}, ignore_index = True)
        except:
            if(keys != 'Parameter' and keys != 'Value' and keys != 'Class'):
                d[keys] = d[keys].append({keys: 'NA'}, ignore_index = True)
    return d
        
d1 = {'a': pd.DataFrame(columns = ['a']), 'b': pd.DataFrame(columns = ['b']), 'c': pd.DataFrame(columns = ['c']), 'Parameter': pd.DataFrame(columns = ['Parameter']), 'Value': pd.DataFrame(columns = ['Value'])}
tempd1 = {'a': 1, 'b': 2}
fill_cols(d1, tempd1)['Value']
```

```python
l = [5, 6, 3, 3, 5, 5, 3]
col = {}
col['Class'] = pd.DataFrame(columns = ['Class'])

for i in range(len(wb.sheet_names())):
    sheet = wb.sheet_by_index(i)
    for j in range(l[i]):
        col[sheet.cell_value(1,j)] = pd.DataFrame(columns = [sheet.cell_value(1,j)])
    col['Parameter'] = pd.DataFrame(columns = ['Parameter'])
    col['Value'] = pd.DataFrame(columns = ['Value'])    
    
for i in range(len(wb.sheet_names())):
    sheet = wb.sheet_by_index(i)
    print('Sheet Number: ' + str(i))
    print('Sheet Name: ' + wb.sheet_names()[i])
    
    for j in range(2, sheet.nrows):
        print(str(j) + ' out of ' + str(sheet.nrows))
        tempcol = {}
        for k in range(l[i]):
            tempcol[sheet.cell_value(1, k)] = sheet.cell_value(j, k)
        
        for k in range(l[i], sheet.ncols):
#             print("Start")
            col = fill_cols(col, tempcol)
#             print("End")
            col['Parameter'] = col['Parameter'].append({'Parameter': sheet.cell_value(1,k)}, ignore_index = True)
            col['Value'] = col['Value'].append({'Value': sheet.cell_value(j,k)}, ignore_index = True)
            col['Class'] = col['Class'].append({'Class': wb.sheet_names()[i]}, ignore_index = True)

df = pd.DataFrame()
for keys in col:
    df = pd.concat((df, col[keys]), axis = 1)
    
df
```
